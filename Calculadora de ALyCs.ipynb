```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Supuestos
weekly_working_hours = 160
avg_wage = 40000
avg_overhead = 1.6
avg_comission = 0.007
avg_rotation = 1.3
minutes_on_an_hour = 60
months_on_a_year = 12
```

```python
#Variables de la ALYC - Unicornie
quantity_customers = 1500 #cantidad de clientes de la alyc
quantity_financial_advisors = 9 #cantidad de asesores financieros o empleados de atención de la alyc
aum = 1500000000 #aum
```

```python
# Variables a calcular
monthly_availability = quantity_financial_advisors * weekly_working_hours * minutes_on_an_hour / quantity_customers 
# minutos al mes disponibles de asesores por cada cliente
monthly_customer_service_unitary_cost = quantity_financial_advisors * avg_wage * avg_overhead / quantity_customers
# costo mensual unitario de atención al cliente
required_networth = (monthly_customer_service_unitary_cost * months_on_a_year) / (avg_comission * avg_rotation)
# patrimonio mínimo requerido para cubrir costos de atención
avg_aum = aum / quantity_customers
required_aum = required_networth * quantity_customers
```

```python
print(monthly_availability, monthly_customer_service_unitary_cost, required_networth, required_aum, avg_aum)
```

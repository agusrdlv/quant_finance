```python
def d_returns(fci, date_from=None, date_until=None):
    daily_returns = fci.loc[date_from:date_until].pct_change()
    daily_returns.dropna(inplace=True)
    return daily_returns                            
    print(daily_returns)
```

```python
def d_log_returns(fci, date_from=None, date_until=None):
    daily_log_returns = np.log(fci.loc[date_from:date_until])-np.log(fci.loc[date_from:date_until].shift(1))
    daily_log_returns.dropna(inplace=True)
    return daily_log_returns                            
    print(daily_returns)
```

```python
since="2019-01-02"
until="2019-07-26"
```

```python
for fci in fcis:
    name = "r_" + fci
    df[name] = d_returns(df[fci], since, until)
```

```python
r_s = ['r_1810 Renta Variable Argentina', 'r_1822 Raices Valores Negociables',
       'r_Alpha Acciones - Clase A', 'r_Arpenta Acciones Argentina',
       'r_Bull Market Acciones Argentinas - Clase A',
       'r_Fima PB Acciones - Clase A', 'r_Gainvest Renta Variable',
       'r_HF Acciones Lideres - Clase G', 'r_Lombard Acciones Lideres',
       'r_Pellegrini Acciones - Clase A', 'r_Pionero Acciones',
       'r_Premier Renta Variable - Clase A',
       'r_SBS Acciones Argentina - Clase A',
       'r_Superfondo Renta Variable - Clase A']
```

```python
volatilities = dict()

for rs in r_s:
    vn = "volatility_"+rs
    v = np.round(df[rs].std() * np.sqrt(len(df[rs])), 4)
    volatilities.update({vn : v})
print (volatilities)
```

```python
def tracking_error(fci, index, date_from=None, date_until=None):
    r_fci = d_returns(fci, date_from, date_until)
    r_index = d_returns(index, date_from, date_until)
    dif = np.power(r_fci - r_index,2)
    dif = dif.dropna()
    d_te =(np.sqrt(sum(dif)/(len(dif)-1))*np.sqrt(len(dif)))
    return d_te
```

```python
tes = dict()

for fci in fcis:
    te_n = "te_"+fci
    te = tracking_error(df[fci], df["Merval"], since, until)
    tes.update({te_n : te})
print (tes)
```

```python
#defino el ratio fci como el exceso de rendimiento respecto del benchmark sobre el tracking error
def fci_ratio(fci, index, date_from=None, date_until=None):
    re = (df.loc[date_until, fci]/df.loc[date_from, fci]-1) - (df.loc[date_until, index]/df.loc[date_from, index]-1)
    te = tracking_error(df[fci], df[index], date_from, date_until)
    return re/te
```

```python
fci_ratio("1810 Renta Variable Argentina", "Merval", date_from="2019-01-02", date_until="2019-07-15")
```

```python
fci_ratios = dict()
for fci in fcis:
    fci_r_n= "fci_ratio"+fci
    fci_r = fci_ratio(fci, "Merval", date_from="2019-01-02", date_until="2019-07-15")
    fci_ratios.update({fci_r_n:fci_r})
print(fci_ratios)
    
```
